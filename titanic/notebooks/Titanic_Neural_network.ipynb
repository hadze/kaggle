# Titanic dataset with Neural Network

### Import libraries

```python
import pandas as pd
```

```python
import warnings
warnings.filterwarnings('ignore')
```

### Load preprocessed data

```python
train_data = pd.read_csv('data/titanic_train_preprocessed.csv')
test_data = pd.read_csv('data/titanic_test_preprocessed.csv')
```

```python
print(f'Train data shape: {train_data.shape}')
print(f'Test data shape: {test_data.shape}')
```

```python
train_data.head()
```

```python
X_train, Y_train = train_data, train_data.Survived
X_train.drop("Survived", axis=1, inplace=True) # Label is not needed in NN
X_test = test_data
```

```python
print(f'Train data X shape: {X_train.shape}')
print(f'Train data Y shape: {Y_train.shape}')
```

```python
X_train.head()
```

### Perform algorithms

#### Neural network

```python
import tensorflow as tf
import keras 
from keras.layers import Dense, Dropout, Input
from keras.models import Sequential
```

#### Define model

```python
model = Sequential()
model.add(Dense(units = 32, input_shape = (7,), activation = 'relu'))
model.add(Dense(units = 128, activation = 'relu'))
model.add(Dense(units = 256, activation = 'relu'))
model.add(Dropout(0.1))
model.add(Dense(units = 1024, activation = 'relu'))
model.add(Dropout(0.1))
model.add(Dense(units = 256, activation = 'relu'))
model.add(Dropout(0.1))
model.add(Dense(units = 32, activation = 'relu'))
model.add(Dense(units =1 , activation = 'sigmoid'))
```

```python
model.summary()
```

```python
_optimizer = 'adam'
_loss = tf.keras.losses.binary_crossentropy
_metrics = (['acc'])
_epochs = 200
_batch_size = 32

model.compile(loss=_loss,optimizer=_optimizer,metrics=_metrics)
```

```python
model.fit(X_train, Y_train, epochs=_epochs) #batch_size=_batch_size
```

#### Prediction on test data

```python
x_test = X_test.drop(['PassengerId'], axis=1)
predict = model.predict(x_test)
#since we have use sigmoid activation function in output layer
Y_pred = (predict > 0.5).astype(int).ravel()
print(predict)
```

### Prepare CSV for submission

```python
test_data.head()
```

```python
submission = pd.DataFrame({
        "PassengerId": X_test["PassengerId"],
        "Survived": Y_pred
    })
submission.to_csv('data/titanic_nn.csv', index=False)
```
